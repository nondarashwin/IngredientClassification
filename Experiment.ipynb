```python
data=pd.read_csv('data.csv')
data.ingredients
```

```python
ingredients=data['ingredients']
cuisine=data['cuisine']
clusters=len(ingredients)
data_model=[]
for i in range(len(cuisine)):  
    a=data['ingredients'][i]
    a=a[1:-1]
    c=a.split(',')
    #c_strip=[]
    for j in c:
        b=j.replace(" ","")
        b=b.strip('\'')
        b=b.lower()
        print(b)
        if(b!=""):
            data_model.append([b,cuisine[i]])
```

```python
data_model=pd.DataFrame(data_model,columns=['Ingrdient','Cuisine'])
data_model
```

```python
from difflib import get_close_matches
types=list(set(data_model['Ingrdient']))
for i in types:
    print(get_close_matches(i,types))
```

```python
le = preprocessing.LabelEncoder()
le.fit(data_model.Ingrdient)
data_model['Ingrdient']=le.transform(data_model['Ingrdient'])
#le1 = preprocessing.LabelEncoder()
#le1.fit(data_model.Cuisine)
#data_model['Cuisine']=le1.transform(data_model['Cuisine'])
```
